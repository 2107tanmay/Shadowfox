```python
#imports
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
```

```python
model_name = "gpt2"  #for now the model is was taken as gpt2 to understand llm from the most basic view point

tokenizer = GPT2Tokenizer.from_pretrained(model_name) #helps us create tokens to convert text to numbers that can be understood by model
model = GPT2LMHeadModel.from_pretrained(model_name) #this defines the actual lm model to be used by us

model.eval() #as model is pretrained due to research purpose and time crunch we are just testing its performance metrics
```

The following code block defines a function that will tokenize our given text and generate text to answer/complete the prompt using pretrained/pre-existing parameters of GPT2 model.

```python
def generate_text(prompt, max_length=10, temperature=0.5, top_k=50, top_p=0.9):
    """
    Generate text using GPT-2 model from a prompt.

    Parameters:
        prompt (str): Input text to condition generation.
        max_length (int): Maximum length of generated sequence.
        temperature (float): Controls randomness of predictions.
        top_k (int): Limits the sampling pool to top k tokens.
        top_p (float): Nucleus sampling probability threshold.

    Returns:
        str: Generated text.
    """

    input_ids = tokenizer.encode(prompt, return_tensors="pt") # encode the text to tokens

    output_ids = model.generate(
        input_ids,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id,
    ) # this will generate the token outputs

    generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    return generated_text
```

The following code block contains a list of strings i.e. prompts that will be given to the model to generate the answer/continuation for the prompts.

```python
prompts = [
    "The future of artificial intelligence is",
    "Gpt2 was trained on ",
    "Financial crash of 2008 left",
    "Capabilities of GPT2 model"
]

for i, prompt in enumerate(prompts):
    print(f"\nPrompt {i+1}: {prompt}")
    print("Generated Text:")
    print(generate_text(prompt))
    print("-" * 80)
```

As we could percieve that using a variation of temperature i.e. sensitivity settings we could alternate the way the model will react to the given list of prompts.

max_lenght: it defines the generated text lengh that's it.

top_k:
it is sampling technique which uses a probability based tokenization system to predict what next token of text might be(the next predicted value of the prompt), a smaller "k" value means that the model is more focused i.e. no randome outcomes may be viewed as we witnessed earlier

top_p:
it is nucleas based sampling technique i.e. it defies the default k values in order to understand the values of "k" and based on that dynamically interact with token. This provides a flexible/dynamic appraoch to the model.

EXAMPLES:

top_k=50 means pick from the 50 most likely tokens only.

top_p=0.9 means pick from the smallest set of tokens that together account for 90% of the total probability mass.

```python
from transformers import GPT2Tokenizer, GPT2Model
from bertviz import head_view

# Load GPT-2 with attention outputs enabled
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2Model.from_pretrained("gpt2", output_attentions=True)

sentence = "The future of artificial intelligence is"
inputs = tokenizer(sentence, return_tensors='pt')

# Forward pass with attention
outputs = model(**inputs)

# Now visualize
head_view(outputs.attentions, sentence.split())
```

In the above code we can see a attention cloud based visualization of our prompt and how its being tokenized as well as how the model will later try to make sense of each token by using a cross reference of tokens of the whole prompt.

This is achieved by breaking down words into smaller parts i.e. subword units.
Attention visualization shows how each token will interact with other tokens at various levels.
